```python
from framenet.ecg.generation import build_cxns_for_frame
from framenet.builder import build
from framenet.valence import get_valence_patterns

fn, fnb = build()

frame = 'Cause_motion'

all_cxns = build_cxns_for_frame(frame, fn, fnb, role_name="Manner", pos="V")
```

## Valences

```python
len(get_valence_patterns(fn.get_frame(frame)))
```

```python
len(all_cxns['cxns_all'])
```

## Number of different `PT` patterns, regardless of order

```python
def t(cxn):
    return tuple(sorted([c.pt.lower() for c in cxn.constituents]))

len(set([t(cxn) for cxn in all_cxns['cxns_all']]))
```
